### Developing an Intuitive Indicator for Data Reliability

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the autocorrelation results CSV file
autocorr_df = pd.read_csv('autocorrelation_results.csv')

# Filter the data for Ground_Truth (SNR_dB = inf)
ground_truth_df = autocorr_df[autocorr_df['SNR_dB'] == float('inf')]

# Extract lag columns
lags = [col for col in ground_truth_df.columns if col.startswith('Lag_')]
lag_values = ground_truth_df[lags].values

# Calculate the mean autocorrelation value for each lag across all curves with SNR = inf
mean_autocorr = lag_values.mean(axis=0)

print("Figure 8(A) Trend of Autocorrelation Function Before Dwell Time.")
# Plot the mean autocorrelation against lag
plt.figure(figsize=(7, 5))
plt.plot(range(1, len(mean_autocorr) + 1), mean_autocorr, marker='o', linestyle='-', label='Mean Autocorrelation (SNR = inf)')

plt.xlabel('Lag')
plt.ylabel('Mean Autocorrelation')
# plt.title('Mean Autocorrelation Function vs Lag (SNR = inf)')
plt.legend(fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the autocorrelation results CSV file
autocorr_df = pd.read_csv('autocorrelation_results.csv')

# Calculate the slopes between consecutive lags for each curve
lag_columns = [col for col in autocorr_df.columns if col.startswith('Lag_')]
lag_values = autocorr_df[lag_columns].values
slopes = np.diff(lag_values, axis=1)

# Calculate the mean slope for each lag
mean_slopes = slopes.mean(axis=0)

# Plot the mean slopes with emphasis on Lag=0 to Lag=1
plt.figure(figsize=(7, 5))
lags = np.arange(1, len(mean_slopes) + 1)

print("Figure 8(B) Slope Between Lag 0 and Lag 1 as a Reliability Metric.")
# Plot all slopes
plt.bar(lags, mean_slopes, color='blue', alpha=0.7, label='Mean Slope between Lags')

# Highlight the slope between Lag=0 and Lag=1
plt.bar(1, mean_slopes[0], color='red', label='Lag=0 to Lag=1')

# Annotate the largest slope change
plt.annotate('Largest Slope Change', 
             xy=(1, mean_slopes[0]), 
             xytext=(3, mean_slopes[0] - 0.05),
             arrowprops=dict(facecolor='red', shrink=0.05),
             fontsize=12, color='red')

plt.xlabel('Lag')
plt.ylabel('Mean Slope')
# plt.title('Mean Slope of Autocorrelation Function vs Lag (Ground Truth, SNR = inf)')
plt.legend(fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### SNR Experiment

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the autocorrelation results CSV file
autocorr_df = pd.read_csv('autocorrelation_results.csv')

# Calculate 1 - Lag_1
autocorr_df['1_Lag_1'] = 1 - autocorr_df['Lag_1']

# Calculate the mean of 1 - Lag_1 for each SNR level
mean_1_lag_1_per_snr = autocorr_df.groupby('SNR_dB')['1_Lag_1'].mean().reset_index()

# Separate Ground_Truth (SNR_dB = inf)
ground_truth = mean_1_lag_1_per_snr[mean_1_lag_1_per_snr['SNR_dB'] == float('inf')]['1_Lag_1'].values[0]
mean_1_lag_1_per_snr = mean_1_lag_1_per_snr[mean_1_lag_1_per_snr['SNR_dB'] != float('inf')]

print("Figure 12: Difference Between Lag 0 and Lag 1 at Various SNR Levels")
# Plot the results
plt.figure(figsize=(14, 4))
plt.plot(mean_1_lag_1_per_snr['SNR_dB'], mean_1_lag_1_per_snr['1_Lag_1'], marker='o', label='Mean of (Lag0 - Lag_1)')

# Plot the Ground_Truth as a horizontal line
plt.axhline(y=ground_truth, color='black', linestyle='--', linewidth=1.25, label='Stable Point')

# Annotate the Ground_Truth value
plt.text(mean_1_lag_1_per_snr['SNR_dB'].max(), ground_truth, f'Stable Point: {ground_truth:.4f}',
         verticalalignment='bottom', horizontalalignment='right', fontsize=14, color='black')

plt.xlabel('SNR (dB)')
plt.ylabel('Mean of (Lag0 - Lag_1)')
# plt.title('Mean of (1 - Lag_1) vs SNR')
plt.legend(fontsize=14)

# Set y-axis interval to 0.01
plt.yticks(np.arange(0, 0.8, 0.1))

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Time Trace Length Experiment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Load the autocorrelation results CSV file
autocorr_df = pd.read_csv('extended_autocorrelation_results.csv')

# Calculate 1 - Lag_1 for each row
autocorr_df['1_Lag_1'] = 1 - autocorr_df['Lag_1']

# Calculate the mean of 1 - Lag_1 for each time trace length
mean_1_lag_1_per_length = autocorr_df.groupby('Length')['1_Lag_1'].mean().reset_index()

# Extract the Length and 1_Lag_1 values
lengths = mean_1_lag_1_per_length['Length']
mean_1_lag_1 = mean_1_lag_1_per_length['1_Lag_1']

# Calculate the relative change in mean of (1 - Lag_1)
relative_change = mean_1_lag_1_per_length['1_Lag_1'].pct_change().abs()

# Define the threshold for relative change (e.g., 3%)
threshold = 0.02

# Find the first length where the relative change is below the threshold
stable_length_index = relative_change[relative_change < threshold].index[0]
stable_length = lengths.iloc[stable_length_index]

print ("Figure 13.  Difference Between Lag 0 and Lag 1 at Various Time Teace Lengths.")
# Plot the mean of (1 - Lag_1) vs Time Trace Length with stable point
fig, ax1 = plt.subplots(figsize=(14, 4))

ax1.plot(lengths, mean_1_lag_1, marker='o', label='Mean of (Lag0 - Lag_1)', color='b')
ax1.axvline(x=stable_length, color='r', linestyle='--', label=f'Stable Point = {stable_length}')
ax1.set_xlabel('Time Trace Length')
ax1.set_ylabel('Mean of (Lag0 - Lag_1)', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.xaxis.set_major_locator(MultipleLocator(20))

# fig.suptitle('Mean of (1 - Lag_1) vs Time Trace Length')
fig.tight_layout()  # to prevent overlap of y-axis labels

ax1.legend(loc='upper right', fontsize=14)
ax1.grid(True)

plt.show()
```

### Application of the Reliability Indicator 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load a new dataset from the CSV file
input_file = 'autocorrelation_results1.csv'
data_df = pd.read_csv(input_file)

# Calculate the difference between lag_0 and lag_1 directly from the data
data_df['Difference'] = data_df['Lag_0'] - data_df['Lag_1']

# Determine pass or fail based on the difference
data_df['Pass'] = np.where(data_df['Difference'] > 0.1, 0, 1)  # 0 for FAIL, 1 for PASS

# Group by 'SNR_dB' and calculate the pass ratio
pass_ratio_df = data_df.groupby('SNR_dB')['Pass'].mean().reset_index()

print("Figure 14. Pass Ratio Across Varying SNR Levels")
# Plot the pass ratio as a line plot
plt.figure(figsize=(14, 5))
plt.plot(pass_ratio_df['SNR_dB'], pass_ratio_df['Pass'], marker='o', linestyle='-', color='b')
# plt.title('Pass Ratio vs. SNR_dB')
plt.xlabel('SNR (dB)')
plt.ylabel('Pass Ratio')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(True)
plt.show()
```
